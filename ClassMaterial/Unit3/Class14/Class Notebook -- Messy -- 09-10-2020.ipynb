```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
```

```python
df = pd.read_csv('../data/iowa_train2.csv')
```

```python
df.head()
```

```python
missing_cols_query = df.isnull().sum() > 0
missing_cols_num = df.loc[:, missing_cols_query].select_dtypes(include=np.number).columns.tolist()
missing_cols_cat = df.loc[:, missing_cols_query].select_dtypes(include=np.object).columns.tolist()
df[missing_cols_num] = df[missing_cols_num].fillna(0)
df[missing_cols_cat] = df[missing_cols_cat].fillna('None')
```

```python
# declare X & y
X = df.drop('SalePrice', axis=1)
y = df['SalePrice']

#X = pd.get_dummies(X)
```

```python
X.head()
```

```python
gbm = GradientBoostingRegressor()
gbm.fit(X, y)
```

```python
gbm.score(X, y)
```

```python
# split your data into training & test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1985)
```

```python
max_depth = [2, 3, 4, 5]
learning_rate = [.001, .01, .1]
num_trees = [100, 500, 1000]
cv_scores = []

for depth in max_depth:
    for rate in learning_rate:
        for tree in num_trees:
            print(f"Testing model with: depth: {depth}, rate: {rate}, trees: {tree}")
            pipe.steps[1][1].set_params(max_depth=depth, learning_rate=rate, n_estimators=tree)
            scores = cross_val_score(estimator=pipe, X=X_train, y=y_train, cv=5)
            cv_scores.append((scores.mean(), depth, rate, tree))
```
