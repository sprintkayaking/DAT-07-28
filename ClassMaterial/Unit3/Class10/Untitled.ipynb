```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/housing.csv')
```

```python
df.head()
```

```python
def average_accuracy(df):
    prediction = df['PRICE'].mean()
    avg_accuracy = np.mean(np.abs(df['PRICE'] - prediction))
    return avg_accuracy
```

```python
np.mean(np.abs(df['PRICE'] - df['PRICE'].mean()))
```
